# Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
churn_data = pd.read_csv('Data/churn_data.csv')
churn_data
```

# Data Exploration

```python
churn_data.info()
```

No missing or null values to worry about.

```python
churn_data.describe()
```

```python
for col in churn_data.columns:
    print(churn_data[col].value_counts())
```

I decided to look at specifically the customers who stopped using the service.

```python
churn_customers = churn_data[churn_data['churn'] == True]
churn_customers
```

```python
for col in churn_customers.columns:
    print(churn_customers[col].value_counts())
```

```python
churn_customers.describe()
```

```python
print(churn_data.groupby(['churn']).mean())
```

```python
sns.boxplot(x='churn', y = 'account length', data = churn_data, sym = '')
plt.show()
```

No significant difference between customers who didn't churn or did based on account length

```python
sns.boxplot(x='churn', y = 'customer service calls', data = churn_data, sym = '')
plt.show()
```

Significant difference between customer groups based on number of customer service calls. On average, customers who churned made more customer service calls.

```python
sns.boxplot(x='churn', y = 'total day charge', data = churn_data, sym = '')
plt.title('Customer Churn vs Total Day Charge')
plt.xlabel('Churn')
plt.ylabel('Total Day Charge')
plt.show()
```

Customers who churned had a higher day charge on average compared to customers who stayed.
